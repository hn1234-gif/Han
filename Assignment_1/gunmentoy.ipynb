## Rock Paper Scissors game

```python
"""
A simple rock-paper-scissors game function.
"""

# ask user for input
def user_input():
    player1 = input("Player 1, enter your choice (rock, paper, or scissors): ").lower()
    player2 = input("Player 2, enter your choice (rock, paper, or scissors): ").lower()
    return player1, player2

def rps_game(player1, player2):
    """
    Simulates a rock-paper-scissors game between two players.

    Args:
        player1 (str): The choice of player 1 ('rock', 'paper', or 'scissors').
        player2 (str): The choice of player 2 ('rock', 'paper', or 'scissors'). 
    Returns:
        str: The result of the game ('Player 1 wins', 'Player 2 wins', or 'Tie').
    """
    choices = ['rock', 'paper', 'scissors']
    
    if player1 not in choices or player2 not in choices:
        return "Invalid input! Please choose 'rock', 'paper', or 'scissors'."
    
    if player1 == player2:
        return "Tie"
    
    if (player1 == 'rock' and player2 == 'scissors') or \
      (player1 == 'paper' and player2 == 'rock') or \
      (player1 == 'scissors' and player2 == 'paper'):
        return "Player 1 wins"
    else:
        return "Player 2 wins" 

def main():
    player1, player2 = user_input()
    result = rps_game(player1, player2)
    print(result)
```
